### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    #print(d.find_all('em')[-1].text)
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
clsList=[]

for d in soup.find(class_="part_list_2").find_all('h3'):
    sdic = dict(cls=d.find_all('em')[-1].text, dt= d.find(class_="date").text, vlu = d.find_all('a')[-1].text)
    clsList.append(sdic)
    
clsList
```

### ③ 哪一個類別的文章最多

```python
sList = []
title = []
for scls in clsList:
    #print(scls)
#    for key in scls:
#        print(key)
#    print(scls['cls'])
    #print(sdic.get(scls['cls'],'N'))
    sList.append(scls['cls'])
    ex = False
    for t in title:
        if (ex == False):
            if (t == scls['cls']):
                ex = True
    if ex == False:
        title.append(scls['cls'])
title                

for t in title:
    print(t, sList.count(t))



    
    
```
