```python

from selenium import webdriver
from bs4 import BeautifulSoup

browser = webdriver.Chrome(executable_path='/home/insidehome/2nd-PyCrawlerMarathon/homework/chromedriver')

browser.get("https://www.ettoday.net/news/news-list.htm")
```

```python
# 每個兩秒鐘自動往下滑

import time
for i in range(10):
    time.sleep(2)
    browser.execute_script("window.scrollTo(0, 10000);")
```

```python
# 取得資料，丟到 BeautifulSoup 解析

html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

```python
# 關閉瀏覽器
browser.quit();
```

## 作業目標

根據範例：

1. 取出今天所有的新聞
2. 取出現在時間兩小時內的新聞
3. 根據範例，取出三天前下午三點到五點的新聞

## 1. 取出今天所有的新聞

```python
import datetime

today  = datetime.datetime.now()
print(today)
today_str = today.strftime("%Y/%m/%d")
print(today_str)

for d in soup.find(class_="part_list_2").find_all('h3'):
    if (today_str in d.find(class_="date").text):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 2. 取出現在時間兩小時內的新聞

```python
from datetime import timedelta 
latestDT = (today - timedelta(minutes=120))
today_str = latestDT.strftime("%Y/%m/%d %H:%m")
print(today_str)
for d in soup.find(class_="part_list_2").find_all('h3'):
    if (d.find(class_="date").text > today_str):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
from datetime import timedelta 
latestDT = (today - timedelta(days=3))
target1 = latestDT.strftime("%Y/%m/%d 15:00")
target2 = latestDT.strftime("%Y/%m/%d 17:00")
#target1 = today.strftime("%Y/%m/%d 09:00")
#target2 = today.strftime("%Y/%m/%d 09:28")
print(target1, ":", target2)

for d in soup.find(class_="part_list_2").find_all('h3'):
    if ((d.find(class_="date").text >= target1) and (d.find(class_="date").text <= target2)):
        print(d.find(class_="date").text, d.find_all('a')[-1].text)
```
