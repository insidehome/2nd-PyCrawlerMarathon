### 比較一下範例檔案中的「File I/O」與「xmltodict」讀出來的內容有什麼差異


```python
'''
Your Code 一個是字串，另一個可以JSon陣列方式取得資料
'''
# 讀檔案
fh = open("../homework/data/Day003/64_72hr_CH.xml", "r")
xml = fh.read()
fh.close()

#print("File I/O: ", xml)

# 解析檔案內容
import xmltodict
d = dict(xmltodict.parse(xml))
#print("xmltodict: ", d)
```

### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

'''
Your Code
'''
# 取出 datasetDescription
datasetDescription = d['cwbopendata']['dataset']['datasetInfo']['datasetDescription']
#print(datasetDescription)
datasetDescription = d['cwbopendata']['dataset']['locations']['location']
#print(datasetDescription)

    
for local in datasetDescription:
    location = local["locationName"]
    #print(location)
    elements = local["weatherElement"]
    hasTemp = False
    for element in elements:
        tempFlag = element["description"]
        #print(tempFlag)
        if (tempFlag == "溫度"):
            hasTemp = True
            print(location)
        
        
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

'''
Your Code
'''
for local in datasetDescription:
    dataStr = ""
    location = local["locationName"]
    #print(location)
    elements = local["weatherElement"]
    for element in elements:
        tempFlag = element["description"]
        #print(tempFlag)
        if (tempFlag == "溫度"):
            dtimes = element["time"][1]
            print(location + "," + dtimes["dataTime"], ",", dtimes["elementValue"]["value"])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度

'''
Your Code
'''
for local in datasetDescription:
    dataStr = ""
    location = local["locationName"]
    #print(location)
    elements = local["weatherElement"]
    for element in elements:
        tempFlag = element["description"]
        #print(tempFlag)
        if (tempFlag == "溫度"):
            dtimes = element["time"]
            for dt in dtimes:
                temps = dt["elementValue"]["value"]
                #print(temps)
                dataStr = (location + "," + dt["dataTime"] + "," + temps)
                print(dataStr)
```
