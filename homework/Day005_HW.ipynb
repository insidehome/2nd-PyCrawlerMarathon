## 作業目標

* 請利用 API: https://www.dcard.tw/_api/forums/pet/posts?popular=true 回答下列問題：

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？
import requests, json 
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')
response = r.text
#import json
data = json.loads(response)
data

i = 0
j = 0
for d in data:
    i +=  1
    j = 0
    for col in d.keys():
        j += 1
        print('欄位名稱: ', col)
print('ㄧ次傳回的筆數: ' , i , ' 欄位數有: ', j)
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」

for d in data:
    print('標題： ', d['title'], ', 貼文時間： ', d['createdAt'] , ', 留言人數： ', d['commentCount'] , ', 按讚人數： ', d['likeCount'])
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
rowCount = 0
replyCount = 0
likeCount = 0
for d in data:
    rowCount +=  1
    replyCount += d['commentCount']
    likeCount += d['likeCount']
print('熱門文章有: ' , rowCount , ', 「平均留言人數」: ', (replyCount/rowCount), ', 「平均按讚人數」: ', (likeCount/rowCount))


    
```

```python
r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
response = r.text
#import json
data = json.loads(response)
rowCount = 0
replyCount = 0
likeCount = 0
for d in data:
    rowCount +=  1
    replyCount += d['commentCount']
    likeCount += d['likeCount']
print('非熱門文章有: ' , rowCount , ', 「平均留言人數」: ', (replyCount/rowCount), ', 「平均按讚人數」: ', (likeCount/rowCount))
```
